```python
velocidades = []

for i in range(5):
    velocidad = int(input("Ingrese una velocidad: "))
    velocidades.append(velocidad)

suma = 0

for velocidad in velocidades:
    suma = suma + velocidad

promedio = suma / 5

mayor = velocidades[0]

for velocidad in velocidades:
    if velocidad > mayor:
        mayor = velocidad

print("Promedio:", promedio)
print("Velocidad maxima:", mayor)

todas = True

for velocidad in velocidades:
    if velocidad > 60 or velocidad < 120:
        print("Estas dentro del limite")
    if velocidad < 20 or velocidad < 140:
        print("Peligrp")
    

if todas:
    print("Todas las velocidades estan dentro del limite")
```
